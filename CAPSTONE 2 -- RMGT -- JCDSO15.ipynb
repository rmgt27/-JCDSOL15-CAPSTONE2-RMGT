##### 3.1.1 Penanganan ID dengan Nilai Nol Menggunakan ID Acak

Untuk mengatasi ID dengan nilai nol dalam dataset, kita bisa menggantinya dengan angka acak yang unik. Berikut adalah langkah-langkah yang dapat dilakukan:

```python
# Memeriksa apakah ada data dengan ID bernilai nol
df.loc[df['ID'] == 0] #df.loc[df['ID'] == 0] digunakan untuk memeriksa apakah ada ID yang bernilai nol dalam dataset.

# Menghasilkan angka acak untuk menggantikan ID dengan nilai nol
id_list = df['ID'].tolist()
np.random.seed(22)  # Set seed untuk mendapatkan hasil yang sama setiap kali dijalankan
id_0 = 0

# Menghasilkan angka acak yang unik untuk menggantikan ID nol
while True:
    random_number = np.random.randint(low=1, high=10000)
    if random_number not in id_list:
        id_0 = random_number
        break

# Mengganti nilai nol pada ID dengan angka acak yang dihasilkan
df['ID'] = df['ID'].replace(0, id_0)

# Memeriksa apakah penggantian berhasil
df.loc[df['ID'] == id_0]
```

Untuk mengelompokkan umur menjadi kategori sesuai dengan pedoman Kementerian Kesehatan Indonesia, kamu bisa membuat kolom baru Age_categories yang mengelompokkan umur ke dalam kategori yang telah ditentukan. Berikut adalah langkah-langkah dan kode dalam Bahasa Indonesia:

Kategori Usia:

1. Toddler: Kurang dari 5 tahun
2. Children: 5-11 tahun
3. Teenage: 12-25 tahun (Gabungan Early Teenage dan Late Teenage)
4. Early Adult: 26-35 tahun
5. Late Adult: 36-45 tahun
6. Early Elderly: 46-55 tahun
7. Late Elderly: 56-65 tahun
8. Old: Lebih dari 65 tahun


```python
# Membuat fungsi untuk mengelompokkan usia
def labelage(x):
    if x <= 5:
        return '1. Toddler'
    elif x <= 11:
        return '2. Children'
    elif x <= 25 :
        return '3. Teenage'
    elif x <= 45:
        return '4. Adult'
    elif x <= 65 :
        return '5. Elderly'
    else:
        return '6. Old'
    
df.insert(3, column='Age_category', value=df['Age'].apply(labelage))

df.head ()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Asumsikan kita mengklaster berdasarkan pendapatan dan total pembelian anggur
X = data[['Income', 'MntWines']]

# Melakukan KMeans Klasternisasi
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
data['Cluster'] = kmeans.labels_

# Visualisasi hasil klaster
plt.figure(figsize=(10, 6))
plt.scatter(data['Income'], data['MntWines'], c=data['Cluster'], cmap='viridis')
plt.colorbar(label='Cluster')
plt.xlabel('Income')
plt.ylabel('MntWines')
plt.title('Segmentasi Pelanggan Berdasarkan Income dan Pembelian Anggur')
plt.show()
```

Kita akan melakukan uji ANOVA untuk melihat apakah terdapat perbedaan signifikan dalam pengeluaran (`Income`) berdasarkan status pernikahan (`Marital_Status`) dan tingkat pendidikan (`Education`).


```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# ANOVA untuk pengeluaran berdasarkan status pernikahan
anova_marital = ols('Income ~ C(Marital_Status)', data=df).fit()
anova_table_marital = sm.stats.anova_lm(anova_marital, typ=2)
anova_table_marital
```

```python

# ANOVA untuk pengeluaran berdasarkan pendidikan
anova_education = ols('Income ~ C(Education)', data=df).fit()
anova_table_education = sm.stats.anova_lm(anova_education, typ=2)
anova_table_education
```


## Analisis Inferensial: Regresi Linear
Selanjutnya, kita akan melakukan analisis regresi untuk mengevaluasi pengaruh `Income`, `Marital_Status`, dan `Education` terhadap respons pelanggan terhadap kampanye pemasaran (`Response`).


```python

# Regresi linear untuk melihat pengaruh pendapatan, status pernikahan, dan pendidikan terhadap respons
df['Response_binary'] = df['Response'].apply(lambda x: 1 if x == 1 else 0)
model = ols('Response_binary ~ Income + C(Marital_Status) + C(Education)', data=df).fit()
model.summary()
```
